```python
with open("alice_in_wonderland.txt","r",encoding="utf-8") as f:
    raw_text = f.read()
print('Charcaters:',len(raw_text))
print(raw_text[:1000])
```

```python
import re

preprocessed = re.split(r'([,.:;?_!"()\]]|--|\s)', raw_text) 
preprocessed = [item for item in preprocessed if item.strip()]
print(len(preprocessed))
```

```python
all_words = sorted(set(preprocessed))
vocab_size = len(all_words)
print(vocab_size)
```

```python
vocab = {token:integer for integer,token in enumerate(all_words)}
for integer,token in enumerate(vocab.items()):
    if integer>100:
        break
    print(f'{token}')
word = next(token for token, idx in vocab.items() if idx == 43)
print(word)
```

```python
from simpleTokenizer import SimpleTokenizer

tokenizer = SimpleTokenizer(vocab)
text = "Alice was beginning to get very tired of sitting by her sister on the bank"
ids = tokenizer.encode(text)
print(ids)
decoded_text = tokenizer.decode(ids)
print(decoded_text)
```

Handling unknown and end_of_text tokens

```python
all_tokens = sorted(list(set(preprocessed)))
all_tokens.extend(["<|unk|>","<|endoftext|>"])
vocab = {token:integer for integer,token in enumerate(all_tokens)}
print(len(vocab.items()))

#Printing last 5 items of the dictionary

for i,item in enumerate(list(vocab.items())[-5:]):
    print(item)
```

```python
text1 = 'Alice was beginning to get very tired of sitting by her sister on the bank'
text2 = 'The sun grew old and the birds flew past the horizon.'
text = " <|endoftext|> ".join([text1,text2])

tokenizer = SimpleTokenizer(vocab)
print('Encoded tokens:')
ids = tokenizer.encode(text)
print(ids)
print('Decoded text:')
decoded_text = tokenizer.decode(ids)
print(decoded_text)
```

Byte Pair Encoding Implementation

```python
import importlib
import tiktoken

print('tiktoken version:',importlib.metadata.version("tiktoken"))
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
text = (
    "Do you like tea? <|endoftext|> In the sunlit terraces"
    "of someunknownPlace."
)
integers = tokenizer.encode(text,allowed_special={"<|endoftext|>"})
print(integers)
```

```python
decoded_text = tokenizer.decode(integers)
print(decoded_text)
```
